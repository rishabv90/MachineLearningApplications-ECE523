```python
import matplotlib.pyplot as pyplot #to draw 
import numpy as np #for array manupilation
from sklearn.datasets import make_moons as mm #make moons dataset
from sklearn.linear_model import LogisticRegression #regression function number 1
from sklearn.ensemble import RandomForestClassifier #regression fucntion number 2
import warnings
warnings.filterwarnings('ignore')
```

```python
#logistic regression solution
a,b = mm(800, noise=0.3, random_state=0)
pyplot.figure()
pyplot.plot(a[np.where(b==1)[0], 0], a[np.where(b==1)[0], 1], 'ro', marker = 's')
pyplot.plot(a[np.where(b==0)[0], 0], a[np.where(b==0)[0], 1], 'go', marker = 's')
pyplot.show()

classifier = LogisticRegression(random_state=0)
classifier.fit(a,b)
a1, b1 = mm(800, noise=0.3, random_state=0)


bhat_rf = classifier.predict(a1)
pyplot.plot(a1[np.where(bhat_rf==1)[0], 0], a1[np.where(bhat_rf==1)[0], 1], 'ro', marker = 's')
pyplot.plot(a1[np.where(bhat_rf==0)[0], 0], a1[np.where(bhat_rf==0)[0], 1], 'go', marker = 's')
pyplot.show()
```

```python
#with random forrest the answer is more accurate. 
a,b = mm(800, noise=0.3, random_state=0)
pyplot.figure()
pyplot.plot(a[np.where(b==1)[0], 0], a[np.where(b==1)[0], 1], 'ro', marker = 's')
pyplot.plot(a[np.where(b==0)[0], 0], a[np.where(b==0)[0], 1], 'go', marker = 's')
pyplot.show()
classifier = RandomForestClassifier(random_state=0)
classifier.fit(a,b)
a1, b1 = mm(5000, noise=0.3, random_state=0)


bhat_rf = classifier.predict(a1)
pyplot.plot(a1[np.where(bhat_rf==1)[0], 0], a1[np.where(bhat_rf==1)[0], 1], 'ro', marker = 's')
pyplot.plot(a1[np.where(bhat_rf==0)[0], 0], a1[np.where(bhat_rf==0)[0], 1], 'go', marker = 's')
pyplot.show()
```
